## What is PASy?

**PASy (Pernage Analytics System)** is a synthetic benchmarking framework
designed for industrial operations that have no comparable peer facilities.

Conventional benchmarking methods — including Data Envelopment Analysis (DEA) —
require multiple comparable units to construct an efficiency frontier.
When only one facility exists, or when peer data is unavailable,
those methods become infeasible.

PASy addresses this by **reconstructing the benchmarking environment**
rather than modifying the DEA model itself. It generates a structured set
of synthetic peer DMUs from a single observed facility's operating data,
then applies bounded DEA to produce efficiency scores, performance rankings,
and quartile-based diagnostic categories.

The framework integrates:
- **PRISM** — a domain classification scheme for operational variables
- **AHP** — Analytic Hierarchy Process weighting of variable importance
- **Beta-distributed perturbation** — bounded stochastic synthetic generation
- **Bounded DEA** — optimistic and pessimistic frontier evaluation
